```python
# Why random values?

# np.empty() does not fill the array with zeros.

# It allocates memory and leaves whatever data was already present in that memory location.

# Think of it as:

# Allocate memory quickly
# Don't clean it
# Use whatever bytes are already there
```

```python
# np.empty(shape,dtype) : Create with random values

empty_arr = np.empty(5)
empty_arr
```
