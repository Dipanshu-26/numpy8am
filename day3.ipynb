```python
# Convert Float Array to Integer
arr_int=npa3.astype(int)
print(arr_int)
print(arr_int.dtype)
```

```python
# Convert Float Array to String
arr_str=arr_int.astype(str)
print(arr_str)
print(arr_str.dtype)
```

```python
# Convert Float Array to float
arr_flt=arr_str.astype(float)
print(arr_flt)
print(arr_flt.dtype)
```
